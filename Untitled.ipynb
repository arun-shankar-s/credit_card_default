```python
# Meaning of PAY_0

# PAY_0 = repayment status in September (most recent month before prediction)

# It tells whether the customer paid on time, delayed payment, or had serious delay.

# PAY_0 Value	Meaning
# -2	No consumption
# -1	Paid in full
# 0	Payment made on time
# 1	Payment delayed 1 month
# 2	Payment delayed 2 months
# 3	Payment delayed 3 months
# 4	Payment delayed 4 months
# 5	Payment delayed 5 months
# 6	Payment delayed 6 months
# 7	Payment delayed 7 months
# 8	Payment delayed 8 months
# 9	Payment delayed 9 months or more


sns.countplot(x='PAY_0', hue='default', data=df)
plt.title("Repayment Status vs Default")
plt.show()


# Blue bars (0) → Clients who did NOT default means the customer successfully pays their credit card due
# Orange bars (1) → Clients who defaulted
```

#### Insight
##### Repayment status has a strong influence on default behavior. Customers who make payments on time (PAY_0 ≤ 0) have a significantly lower probability of default. However, as the repayment delay increases, the number of default cases rises sharply. Clients with delays of two months or more show a substantially higher likelihood of defaulting on their credit card payments.

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

The correlation heatmap reveals strong relationships among repayment status variables (PAY_0–PAY_6) and among billing amount variables (BILL_AMT1–BILL_AMT6), indicating consistent payment behavior and spending patterns across months. Among all features, PAY_0 shows the highest correlation with the target variable, suggesting that recent repayment behavior is the most important indicator of credit default risk.

## Feature Engineering

```python
df = pd.get_dummies(df, columns=['SEX', 'EDUCATION', 'MARRIAGE'], drop_first=True)
```

```python
df['avg_bill_amt'] = (
    df['BILL_AMT1'] +
    df['BILL_AMT2'] +
    df['BILL_AMT3'] +
    df['BILL_AMT4'] +
    df['BILL_AMT5'] +
    df['BILL_AMT6']
) / 6
```

```python
df['avg_pay_amt'] = (
    df['PAY_AMT1'] +
    df['PAY_AMT2'] +
    df['PAY_AMT3'] +
    df['PAY_AMT4'] +
    df['PAY_AMT5'] +
    df['PAY_AMT6']
) / 6
```

```python
df['total_delay'] = (
    df['PAY_0'] +
    df['PAY_2'] +
    df['PAY_3'] +
    df['PAY_4'] +
    df['PAY_5'] +
    df['PAY_6']
)
```

```python
df['credit_utilization'] = df['avg_bill_amt'] / df['LIMIT_BAL']
```

```python
df['payment_ratio'] = df['avg_pay_amt'] / (df['avg_bill_amt'] + 1)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_features = [
'LIMIT_BAL','AGE',
'avg_bill_amt','avg_pay_amt',
'total_delay','credit_utilization',
'payment_ratio'
]

df[numerical_features] = scaler.fit_transform(df[numerical_features])
```
